```python
from splinter import Browser
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup as soup
my_service = Service(executable_path='C:\chromedriver_win32\chromedriver.exe')
browser = Browser('chrome', service=my_service)
```

```python
# Visiting the website
base_url = "https://www.ubereats.com/ca/city/"
location = "calgary-ab"
url = base_url + location
browser.visit(url)
```

# My addition starts here

```python
html = browser.html
soup = soup(html, 'html.parser')
```

```python
# Restaurant name

res_name = soup.find_all('h3', class_ = 'be by bg db b1 br dd c3 j4 j5 j6 j7')
restaurants = []

for i in res_name:
    restaurant = i.text
    restaurants.append(restaurant)
    
print(restaurants)
```

```python
# Categorical information

Category = soup.find_all('span', class_ = 'be bf bz db h1 dd c5 c3 br ja')
for i in Category:
    print(i.text)
```

```python
# Rating

Rating = soup.find_all('div', class_ = 'be j8 bg bh fc fb j9 ef i7 al bm cb')
for i in Rating: 
    print (i.text)
```

```python
# Address

Address = soup.find_all('span', class_ = "be bf bz db h1 dd jc c3 br ja")
for i in Address: 
    print(i.text)
```

# My addition ends here

```python
# Set the URL of the category you want to scrape
base_url2 = "https://www.ubereats.com/ca/category/"
cuisine = "pizza"
url2 = base_url2 + location + cuisine
browser.visit(url2)
```

```python
# Extracting all div elements with the class "ak c3"
html = browser.html
soup = soup(html, 'html.parser')
div_elements = soup.find_all('div', class_='ak c3')
```

```python
# Printing the results
for div in div_elements:
    print(div)
```
